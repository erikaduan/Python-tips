**Exploring vocabulary: Useful functions**  
NLTK makes it really easy to get basic information about the size of a text and the complexity of its vocabulary using python functions.  
+ **len(text1)** gives the number of symbols or 'tokens' in your text. E.g. words  
+ **set(text2)** gives you a list of all the tokens in the text, without the duplicates. Hence, len(set(text3)) will give you the total number unique tokens. Remember this still includes punctuation  
+ **sorted(text4)** places items in the list into alphabetical order, with punctuation symbols and capitalised words first.  

```python
Austen_Pers = gutenberg.words('austen-persuasion.txt')
Austen_Sense = gutenberg.words('austen-sense.txt')
Austen_Emma = gutenberg.words('austen-emma.txt')
Caesar = gutenberg.words('shakespeare-caesar.txt')
Alice = gutenberg.words('carroll-alice.txt')
```

```python
sorted(set(Caesar)) #shows the sort hierarchy i.e. punctuation, numbers, Capitalised words
```

We can investigate the lexical richness of a text. For example, by dividing the total number of words by the number of unique words, we can see the average number of times each word is used.

```python
def textRichness(text):
    textRichness = len(text)/len(set(text))
    print("The lexicon richness of '%s' is %i. This is the average times a unqiue word is repeated in the text." %(text, textRichness))
    return

textRichness(Austen_Sense)
textRichness(Austen_Pers)
textRichness(Austen_Emma)
textRichness(Caesar)
textRichness(Alice)
```

```python
# Counting how many times the word love features in each text

for text in Austen_list:
    love_count = text.count('love')
    all_count = len(text)
    love_percent = (love_count/all_count)*100
    print("For %s, the number of times the word 'love' is used is %f percent." %(text, love_percent))
```

```python
# Writing a function that allows you to count specific word frequencies in each text

def count_word(text, word): 
    word_count = text.count(word)
    all_count = len(text)
    word_percent = (word_count/all_count)*100
    return print("For %s, the number of times the word '%s' is used is %f percent." %(text, word, word_percent))
    

count_word(Austen_Emma, 'Emma') #word needs to be written as a string
```

**Fine-grained selection of words**  
Selection of longer words can be defined via set theory:  
*  **[w for w in V if p(w)]**  
*  w = word  
*  V = vocabulary  
*  p(w) = condition that word fulfills

```python
V = set(Austen_Sense)
long_words = [word for word in V if len(word) > 6]
sorted(long_words)
```

```python
fdist_Sense['a'] #i.e. the word 'a' is found 2043 times. 
```

```python
sorted(w for w in fdist_Sense if len(w) > 5 and fdist_Sense[w] > 50) 
# sorts based on removing short words and words that repeat infrequently
# more reflective of story content
```
